## Get all Data

```python
iso_codes = {}

for index, name in df['Country (according to CN Gov):'].items() :
    if name in iso_codes :
        continue
    r = requests.get('https://restcountries.eu/rest/v2/name/'+name)
    if r.status_code==200 :
        country_code = r.json()[0]['alpha3Code']
        iso_codes[name] = country_code
    else : 
        print('error')
    
```

## Correct errors

```python
for index, name in df['Country (according to CN Gov):'].items() :
    if name in iso_codes :
        continue
    cleaned_name=name.split('(')[0]
    cleaned_name=cleaned_name.strip()
    if name == 'Thailiand' :
        cleaned_name = 'Thailand'
    if name == 'The United Arab Emirates' :
        cleaned_name = 'United Arab Emirates'
    r = requests.get('https://restcountries.eu/rest/v2/name/'+cleaned_name)
    if r.status_code==200 :
        country_code = r.json()[0]['alpha3Code']
        iso_codes[name] = country_code
    else : 
        print('error')
        print(name)
```

## Initialise GC tax list

```python
GC_tax = {}
```

## Get Data for GC tax

```python
for index, iso in df['iso code'].items() :
    if iso in GC_tax :
        continue
    r = requests.get('http://api.worldbank.org/v2/country/'+iso+'/indicator/GC.TAX.IMPT.ZS?format=json')
    i=0
    if r.status_code==200 :
        data=r.json()[1][0]['value']
        while r.json()[1][i]['value']==None and i<len(r.json()[1])-1 :
            i=i+1
            data=r.json()[1][i]['value']
        print(iso,data)
        GC_tax[iso] = data
    else :
        print('error',iso)
    
```

## correct error for India

```python
for index, iso in df['iso code'].items() :
    if iso in GC_tax :
        continue
    cleaned_iso=iso
    if iso=='IOT' : 
        cleaned_iso='IND'
    r = requests.get('http://api.worldbank.org/v2/country/'+cleaned_iso+'/indicator/GC.TAX.IMPT.ZS?format=json')
    i=0
    if r.status_code==200 :
        data=r.json()[1][0]['value']
        while r.json()[1][i]['value']==None and i<len(r.json()[1])-1 :
            i=i+1
            data=r.json()[1][i]['value']
        print(iso,data)
        GC_tax[iso] = data
    else :
        print('error',iso)
```
