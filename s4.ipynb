```python
heritage_scale = {}
heritage_ranking = {}
for indexdf, namedf in df["Country (according to CN Gov):"].items() : 
    cleaned_name=namedf.split('(')[0]
    cleaned_name=cleaned_name.strip()
    if namedf == 'Thailiand' :
        cleaned_name = 'Thailand'
    if namedf == 'The United Arab Emirates' :
        cleaned_name = 'United Arab Emirates'
    if namedf == 'United States of America (Not BRI)' :
        cleaned_name = 'United States'
    if namedf == 'Brunei' :
        cleaned_name = 'Brunei Darussalam'
    if namedf == 'East Timor' :
        cleaned_name = 'Timor-Leste'
    if namedf == 'Kyrgyzstan' :
        cleaned_name = 'Kyrgyz Republic'
    if namedf == 'Myanmar':
        cleaned_name = 'Burma'
    for i in range(len(heritage)) :
        if cleaned_name == heritage["Country Name"][i+1] :
            heritage_scale [namedf] = heritage["2019 Score"][i+1]
            heritage_ranking [namedf] = heritage["World Rank"][i+1]
    if namedf not in heritage_scale : 
        heritage_scale [namedf] = ""
        heritage_ranking [namedf] = ""
        print ('error', namedf)
        
```

```python
df['HF Index of Economic Freedom (Scale: 1-100)'] = df['Country (according to CN Gov):'].apply(lambda x : heritage_scale[x])
df['Heritage Foundation Index of Economic Freedom (Global Ranking)'] = df['Country (according to CN Gov):'].apply(lambda x : heritage_ranking[x])
```

```python
bti = pd.read_csv('C:\\Users\\Enzo\\Downloads\\data-BTI-2006-2018.csv')
bti[bti['year']==2018]
```

```python
import numpy
bti_2018 = bti[bti['year']==2018]
for index, isocode in df.isocode.iteritems():
    cleaned_iso=isocode
    if isocode=='ROU':
        cleaned_iso='ROM'
    if isocode=='IOT':
        cleaned_iso='IND'
    for index2, isocode2 in bti_2018['country_code'].items():
        if cleaned_iso == isocode2 :
            df['Bertelsmann Transformation index (BTI)'].loc[index]=bti_2018['stat_ind'][index2]
            df['Bertelsmann Transformation index (BTI) - Ranking'].loc[index]=bti_2018['rank_stat_ind'][index2]
    if numpy.isnan(df['Bertelsmann Transformation index (BTI) - Ranking'].loc[index]) == True :
        print (isocode)
```

```python
for index, isocode in df.isocode.iteritems():
    cleaned_iso = isocode
    if isocode=='IOT':
        cleaned_iso='IND'
    for index2, isocode2 in CPI['ISO3'].items():
        if cleaned_iso == isocode2 :
            df['TI Corruption Perception Index'].loc[index]=CPI['CPI Score 2018'][index2]
            df['TI Corruption Perception Index - Ranking'].loc[index]=CPI['Rank '][index2]
    if numpy.isnan(df['TI Corruption Perception Index - Ranking'].loc[index]) == True :
        print (isocode)
```
